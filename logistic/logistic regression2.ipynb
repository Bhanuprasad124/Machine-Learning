```python
df_dummay=pd.get_dummies(df.salary)
df_dummay=df_dummay.astype(int)
df_dummay
```

```python
df_need=df[["satisfaction_level","average_montly_hours",'promotion_last_5years','left','salary']]
df_need
```

```python
df_sub=pd.concat([df_need,df_dummay],axis=1)
```

```python
df_sub
```

```python
df_sub=df_sub.drop(['salary'],axis=1)
df_sub
```

```python
x=df_sub.drop(['left'],axis=1)

```

```python
x
```

```python
y=df_sub.left
```

```python
y
```

```python

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
ts=LogisticRegression()
```

```python
ts.fit(x_train,y_train)
```

```python
ts.predict(x_test)
```

```python
ts.score(x_test,y_test)
```
