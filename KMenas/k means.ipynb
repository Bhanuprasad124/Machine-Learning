```python
# plt.scatter(df['Age'],df['Income($)'],c=df['cluster'])
df1=df[df['cluster']==0]
df2=df[df['cluster']==1]
df3=df[df['cluster']==2]
plt.scatter(df1['Age'],df1['Income($)'],c='red')
plt.scatter(df2['Age'],df2['Income($)'],c='green')
plt.scatter(df3['Age'],df3['Income($)'],c='blue')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='yellow',marker='x')

plt.legend()
```

```python
find_k=range(1,10)
```

```python
wcss=[]
```

```python
for i in find_k:
    km=KMeans(n_clusters=i)
    km.fit(df[['Age','Income($)']])
    wcss.append(km.inertia_)
```

```python
plt.plot(find_k,wcss,'bo-')
```
