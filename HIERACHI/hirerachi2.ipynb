```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
dendrogram(linkage(df,method='ward'))
fig = plt.figure(figsize=(250, 100))
```

```python
z=linkage(df,method='ward')
dendrogram(z)
```

```python
agg=AgglomerativeClustering(n_clusters=3,metric='euclidean')
```

```python
agg.fit(df)
```

```python
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'],c=agg.labels_)
plt.legend()
plt.show()
```
