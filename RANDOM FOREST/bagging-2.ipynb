```python
import pandas as pd
```

```python
df=pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder()
```

```python
col=('Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope')
for i in col:
    df[i]=encoder.fit_transform(df[i])
```

```python
df
```

```python
df.ChestPainType.unique()
```

```python
x=df.drop('HeartDisease',axis=1)
```

```python
y=df.HeartDisease
```

```python
#scale the features
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
scaled=scale.fit_transform(x)
```

```python
scaled_df=pd.DataFrame(scaled,columns=x.columns)
```

```python
scaled_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svm=SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
svm.score(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
svm.predict(x_test)
```

```python
svm.score(x_test,y_test)
```

```python
# the above is the score using train test and split method
# now lets try the same using cross validation
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score=cross_val_score(svm,x,y,cv=10)
score
```

```python
score.mean()
```

```python
# from the above using cross val score we got the same value that we got above using the train test split
```

```python
# now lets use bagging technique
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag=BaggingClassifier(
    estimator=SVC(),
    n_estimators=10,
    max_samples=0.8,
    oob_score=True,
    random_state=None,


)
```

```python
bag.fit(x_train,y_train)
```

```python

bag.oob_score_
```

```python
bag.score(x_train,y_train)
```

```python
bag.score(x_test,y_test)
```

```python
score=cross_val_score(bag,x,y,cv=5)
```

```python
score
```

```python
score.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier

bag=BaggingClassifier(
    estimator=RandomForestClassifier(),
    n_estimators=10,
    max_samples=0.8,
    oob_score=True,
    random_state=None,


)
```

```python
bag.fit(x_train,y_train)
bag.score(x_train,y_train)
```

```python
bag.score(x_test,y_test)
```

```python
rf=RandomForestClassifier()
```

```python
score=cross_val_score(rf,x,y,cv=5)
```
